```python
import pandas as pd
import scipy 
from scipy import stats

pwd = '/Volumes/diedrichsen_data$/data/Cerebellum/Pontine7T/atlases/thalamus/indiv_parcellations/MDTB-ses2/MDTB_ses2_DCBC'
df = pd.read_csv(f'{pwd}/T_results_16.tsv', sep='\t')

group_mean = df.loc[df.type == 'group', 'dcbc'].mean()

# Extract data per condition
data_dcbc = df.loc[df.type == 'data', 'dcbc']
data_group_dcbc = df.loc[df.type == 'data and group', 'dcbc']

# --- 1) One-sample t-tests vs. group line ---
t_data_vs_group, p_data_vs_group = stats.ttest_1samp(data_dcbc, popmean=group_mean)
t_datagrp_vs_group, p_datagrp_vs_group = stats.ttest_1samp(data_group_dcbc, popmean=group_mean)

# --- 2) Paired t-test between data and data+group ---
t_data_vs_datagrp, p_data_vs_datagrp = stats.ttest_rel(data_dcbc, data_group_dcbc)

print("\n=== T-Test Results ===")
print(f"Data vs Group line: t = {t_data_vs_group:.3f}, p = {p_data_vs_group:.4f}")
print(f"Data+Group vs Group line: t = {t_datagrp_vs_group:.3f}, p = {p_datagrp_vs_group:.4f}")
print(f"Data vs Data+Group: t = {t_data_vs_datagrp:.3f}, p = {p_data_vs_datagrp:.4f}")
```

```python
for run, df_run in df.groupby('runs'):
    data_dcbc = df_run[df_run.type == 'data']['dcbc']
    data_group_dcbc = df_run[df_run.type == 'data and group']['dcbc']
    
    t_data_vs_datagrp, p_data_vs_datagrp = stats.ttest_rel(data_dcbc, data_group_dcbc)
    print(f"Run {run}: data vs data+group  t = {t_data_vs_datagrp:.3f}, p = {p_data_vs_datagrp:.4f}")
```

```python
import pandas as pd

# Load your data
df = pd.read_csv(f'{pwd}/T_results_16.tsv', sep='\t')

# 1. Compute subject means
subject_means = df.groupby('subject')['dcbc'].mean().rename('subject_mean')
df = df.merge(subject_means, on='subject')

# 2. Compute grand mean
grand_mean = df['dcbc'].mean()

# 3. Apply normalization
df['dcbc_centered'] = df['dcbc'] - df['subject_mean'] + grand_mean

# Optional: sanity check
print("Before normalization:", df['dcbc'].mean())
print("After normalization:", df['dcbc_centered'].mean())

# You can now use df['dcbc_centered'] for plotting or further statistical analysis
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# "data" line
sb.lineplot(
    data=df[df.type=='data'], 
    x='runs', 
    y='dcbc_centered', 
    ci=95, estimator='mean', label='data only')  

# "data and group" line
sb.lineplot(
    data=df[df.type=='data and group'], 
    x='runs', 
    y='dcbc_centered', 
    ci = 95,
    estimator='mean',
    label='data and group'
)

# Group line (single value)
group_mean = df[df.type=='group'].dcbc.mean()
plt.axhline(group_mean, color='black', linestyle='--', label='group')

plt.xlabel('Run')
plt.ylabel('DCBC (centered per subject)')
plt.title('DCBC across runs (within-subject variability)')
plt.legend()
plt.tight_layout()
plt.show()
```
